```python
import numpy as np
import pandas as pd
import ast
```

```python
credit = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
credit.head(1)
movies.head(1)
print(movies.shape)
print(credit.shape)
```

```python
# merging two dataframes
movies = movies.merge(credit,on="title")
print(movies.shape)
```

```python
movies.info()
```

```python
movies = movies[['movie_id','genres','overview','title','cast','crew','keywords']]
movies.head()
```

```python
# checking whether the data contains the null or missing values 
movies.isnull().sum()
```

```python
# droping the data - since only 3 data is missing
movies.dropna(inplace=True)  
```

```python
movies.isnull().sum()
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.iloc[0].cast
# OR
# movies['cast'][0]     both will give the same output 
# this contains the name of the main characters - and i need only first thee characters - main characters
```

```python
def convert_cast(obj):
    l1 = []
    count=0
    for i in ast.literal_eval(obj):
        if(count==3):
            break
        count += 1
        l1.append(i["name"])
    return l1
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies['cast']
```

```python
def convert_crew(obj):
    l3 = []
    for i in ast.literal_eval(obj):
        if(i["job"]=="Director"):
            l3.append(i["name"])
            break
    return l3
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
movies['crew']
```

```python
def convert_keyword(obj):
    l4= []
    for i in ast.literal_eval(obj):
        l4.append(i["name"])
    return l4
```

```python
movies['keywords'] = movies['keywords'].apply(convert_keyword)
movies['keywords'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())   # now we are going to convert this string into list
```

```python
# now this will remove the space between the words of same word 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tag'] = movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords'] + movies['overview'] # adding all 5 rows into one single 
movies = movies[['movie_id','title','tag']]
movies.head()
```

```python
movies['tag']# now we are going to convert this list into string
movies['tag'] = movies['tag'].apply(lambda x:" ".join(x))
# now converting it into lower case - recommended
movies['tag'] = movies['tag'].apply(lambda x:x.lower())
movies['tag'][0]
```

```python
# now going to do text vectorization => Bag of Words technique , there are many more such as - tfidf,word2vec
# now 1st going to remove stop_words - words used in sentence to maintain continuity are removed - to,is,it
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tag']).toarray()
vectors
```

```python
cv.get_feature_names_out()  # all the list of 5000 words, that are used in vectorization
```

```python
# converting the words with same meaning into a root word 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
# e.g.
print(ps.stem("love"))
print(ps.stem('loved'))
print(ps.stem('loved'))
print(ps.stem('activities'))
def stem(txt):
    y = []
    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y) # " ".join(y) converts the list into string
movies['tag'].apply(stem)
```

```python
movies['tag'] = movies['tag'].apply(stem)
print(movies['tag'][0])
# again doing vectorization with new tag
vectors = cv.fit_transform(movies['tag']).toarray()
cv.get_feature_names_out()
```

```python
# now to check the similarity between two vectors we will compare their cosine angles with each 
# greater the cosine distance lesser the simialrity
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(vectors).shape)  # comparing each movie with all movies - 4806
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
# fetching the index/movie_id of the movie
print(movies['movie_id'][movies[movies['title']=="Pirates of the Caribbean: At World's End"].index[0]])

# fetching similarity of that movies with other movies
print(similarity[movies[movies['title']=="Pirates of the Caribbean: At World's End"].index[0]])

# now sorting the similarity into descending order
print(sorted(similarity[movies[movies['title']=="Pirates of the Caribbean: At World's End"].index[0]],reverse=True))

# but here comes the problem of loosing the index positions,
# so we use enumerate function to convert it into a list of tuple  [(index, cosine_similarity)]
print(list(enumerate(similarity[movies[movies['title']=="Pirates of the Caribbean: At World's End"].index[0]])))

# now sorting wrt enumerated_list
sorted(list(enumerate(similarity[1])),reverse=True,key = lambda x:x[1])
```

```python
# function for finding similar movies
def similar_movies(movie):
    index = movies[movies['title']==movie].index[0]
    m_d = similarity[index]
    m_l = sorted(list(enumerate(m_d)), reverse=True, key=lambda x: x[1])[1:11]
    
    for i in m_l:
        print(movies.iloc[i[0]].title)
        
similar_movies("Spider-Man 2")
```
